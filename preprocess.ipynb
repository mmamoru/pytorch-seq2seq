```python
import unicodedata
import re
import time
def normalize(twitter_data):
    dialogue_tweets = []
    for i in range(len(twitter_data)):
        s = "".join(c for c in twitter_data[i] if unicodedata.category(c) != "Mn")
        s = unicodedata.normalize("NFKD", s)
        dialogue_tweets.append(s)
    return dialogue_tweets

def delete_self_reply_pairs(twitter_data):
    dialogue_tweets = []
    for i in range(1, len(twitter_data), 3):
        if (re.match('@', twitter_data[i])!=None and 
            re.match('@', twitter_data[i+1])!=None):
            dialogue_tweets.append(twitter_data[i-1])
            dialogue_tweets.append(twitter_data[i])
            dialogue_tweets.append(twitter_data[i+1])
    return dialogue_tweets

def delete_brackets2(twitter_data):###é ‘å¼µã£ã¦é¡”æ–‡å­—ã‚’æ¶ˆã—ã¦ã‚‹(ã©ã£ã¡ã‹)
    brackets = []
    brackets.append(r'ã€(.+)ã€‘')
    brackets.append(r'ã€Š(.+)ã€‹')
    brackets.append(r'\((.+)\)')
    brackets.append(r'ï¼ˆ(.+)ï¼‰')
    dialogue_tweets = []
    for ele in twitter_data:
        for bracket in brackets:
            ele = re.sub(bracket,'',ele)
        dialogue_tweets.append(ele)
    return dialogue_tweets

def delete_brackets(twitter_data):###é¢å€’ã ã‹ã‚‰ã‚«ãƒƒã‚³ã¯å…¨éƒ¨æ¶ˆã™(ã©ã£ã¡ã‹)
    print(len(twitter_data))
    brackets = r"[ã€ã€‘ã€Šã€‹()ï¼ˆï¼‰]"
    dialogue_tweets = []
    for i in range(0,len(twitter_data),3):
        if (re.search(brackets, twitter_data[i])==None and 
            re.search(brackets, twitter_data[i+1])==None and
            re.search(brackets, twitter_data[i+2])==None):
            dialogue_tweets.append(twitter_data[i])
            dialogue_tweets.append(twitter_data[i+1])
            dialogue_tweets.append(twitter_data[i+2])
    print(len(dialogue_tweets))
    return dialogue_tweets
```

```python
def delete_self_reply(twitter_data):
    tmp = []
    for i in range(0, len(twitter_data),3):
        if re.match('@', twitter_data[i+1])!=None:
            tmp.append(twitter_data[i])
            tmp.append(twitter_data[i+1])
            tmp.append(twitter_data[i+2])
    return tmp

def delete_hashtag(twitter_data):
    tmp = []
    for ele in twitter_data:
        ele = re.sub('#(.*?\s)','',ele)
        ele = re.sub('ï¼ƒ(.*?\s)','',ele)
        tmp.append(ele)
    return tmp
```

```python

def count_hashtag(twitter_data):
    tmp = []
    hashtag = r'[#ï¼ƒ](.*?\s)'
    for i in range(len(twitter_data)):
        if re.search(hashtag,twitter_data[i])!=None:
            tmp.append(i)
    return tmp

def count_end_hashtag(twitter_data):
    tmp = []
    hashtag = r'[#ï¼ƒ](.*?)$'
    for i in range(len(twitter_data)):
        if re.search(hashtag,twitter_data[i])!=None:
            tmp.append(i)
    return tmp
    
```
