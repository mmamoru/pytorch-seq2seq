```python
def delete_self_reply(twitter_data):
    tmp = []
    for i in range(0, len(twitter_data),3):
        if re.match('@', twitter_data[i+1])!=None:
            tmp.append(twitter_data[i])
            tmp.append(twitter_data[i+1])
            tmp.append(twitter_data[i+2])
    return tmp

def delete_hashtag(twitter_data):
    tmp = []
    for ele in twitter_data:
        ele = re.sub('#(.*?\s)','',ele)
        ele = re.sub('＃(.*?\s)','',ele)
        tmp.append(ele)
    return tmp
```

```python

def count_hashtag(twitter_data):
    tmp = []
    hashtag = r'[#＃](.*?\s)'
    for i in range(len(twitter_data)):
        if re.search(hashtag,twitter_data[i])!=None:
            tmp.append(i)
    return tmp

def count_end_hashtag(twitter_data):
    tmp = []
    hashtag = r'[#＃](.*?)$'
    for i in range(len(twitter_data)):
        if re.search(hashtag,twitter_data[i])!=None:
            tmp.append(i)
    return tmp
    
```

```python
h = count_hashtag(rawdata)
he = count_end_hashtag(rawdata)
for i in range(len(rawdata)):
    if i in he:
        print('he',i,rawdata[i])
    if i in h:
        print('h',i,rawdata[i])
```
